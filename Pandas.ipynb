#Tratando os dados Nulos

Existem diversas formas de tratar dados nulos com o Pandas. Algumas das principais formas são:

Remover os dados nulos: É possível remover as linhas ou colunas que possuem valores nulos utilizando o método dropna() . Esse método remove todas as linhas ou colunas que possuem pelo menos um valor nulo.

Preencher os dados nulos: Utilizando o método fillna(), podemos preencher os valores nulos com um valor específico. Além disso, também é possível utilizar argumentos específicos do método fillna() como o method=”ffill” ou method=”bfill” para preencher os valores nulos com o valor anterior ou posterior, respectivamente.

Interpolar os dados nulos: É possível utilizar o método interpolate() para preencher os valores nulos com valores interpolados, ou seja, valores calculados a partir dos valores vizinhos.

```python
df.isnull()
```

```python
df.isnull().sum() # soma a quanntidade de dados nulos em cada coluna
```

axis=0: Refere-se às linhas. A operação é aplicada ao longo das linhas, ou seja, colunas são afetadas.

Exemplo: Se você fizer uma operação como df.mean(axis=0), ela calculará a média de cada coluna (ao longo das linhas).

axis=1: Refere-se às colunas. A operação é aplicada ao longo das colunas, ou seja, linhas são afetadas.

Exemplo: Se você fizer df.mean(axis=1), ela calculará a média de cada linha (ao longo das colunas).

# Criando colunas numéricas


```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio'] # criando a coluna do valor total por mês
```

```python
dados.fillna(0,inplace = True) # Lidando com valores nulos
dados.head()
```

```python
dados['Valor_por_ano'] = (dados['Valor_por_mes'] * 12) + dados['IPTU'] # criando a coluna do valor total por ano
dados.head()
```

#Criando coluna categórica

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' +  dados['Quartos'].astype(str) + ' quarto(s) ' + ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Descricao']
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
dados.head()
```

```python
dados.to_csv('dados_completos.csv', index=False, sep=';')  #Salvar o novo datset em csv
```
